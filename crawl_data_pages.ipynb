```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# num = range(1, 25)

data_1_page = []
all_data = []
for num in range(1,25):
	try: 
		url = f'https://www.scrapethissite.com/pages/forms/?page_num={num}&per_page=25'
		response = requests.get(url)
		response.raise_for_status()
	except requests.exceptions.RequestException as e:
		print(f'URL is not correct {e}')
		exit
	
	soup = BeautifulSoup(response.text, 'html.parser')

	# Ly tn ct
	col = []
	col = soup.find_all('th')
	col_name = []
	for c in col:
		col_name.append(c.get_text(strip=True))
	# print(col_name)

	# Ly d liu 1 page
	teams = soup.find_all('tr', class_='team')
	# print(teams)
	for r in teams:
		name = r.find('td', class_='name').get_text(strip=True)
		# print(name)
		year = r.find('td', class_='year').get_text(strip=True)
		win = r.find('td', class_='wins').get_text(strip=True)
		loss = r.find('td', class_='losses').get_text(strip=True)
		ot_loss = r.find('td', class_='ot-losses').get_text(strip=True)

		pct_tag = r.find('td', class_= lambda x: x and 'pct' in x.split())
		pct = float(pct_tag.get_text(strip=True) or '0')

		gf = r.find('td', class_='gf').get_text(strip=True)
		ga = r.find('td', class_='ga').get_text(strip=True)

		diff_tag = r.find('td', class_=lambda x:x and 'diff' in x.split())
		diff = float(diff_tag.get_text(strip=True) or '0')

		row = [name, year, win, loss, ot_loss, pct, gf, ga, diff]
		data_1_page.append(row)

	# Dn d liu vo 1 list
	
	all_data.extend(data_1_page)

all_data
```

```python
df = pd.DataFrame(data=all_data, columns=col_name)
df.to_csv('football_team.csv')
```
