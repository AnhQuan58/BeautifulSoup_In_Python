```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.scrapethissite.com/pages/forms/'
```

```python
try:
  response = requests.get(url)
  response.raise_for_status()
except requests.exceptions.RequestException as e:
  print(f'URL is not correct {e}')
  
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
# soup
```

```python
cols = soup.find_all('th')
col_name = []
for c in cols:
  col_name.append(c.get_text().strip())
col_name[-1] = 'Texting Success'
col_name
```

Láº¥y 1 page

```python
teams = soup.find_all('tr', class_='team')
data = []
for t in teams:
  name = t.find('td', class_='name').get_text(strip=True)
  # print(name)
  year = t.find('td', class_='year').get_text(strip=True)
  wins = t.find('td', class_='wins').get_text(strip=True)
  losses= t.find('td', class_='losses').get_text(strip=True)
	
  OT_losses = t.find('td', class_='ot-losses').get_text(strip=True)
  if OT_losses == '':
    OT_losses = 0
  # print(OT_losses)
  # win_percen = t.find('td', class_='pct text-success').get_text(strip=True)
  pct_tag = t.find('td', class_=lambda x: x and 'pct' in x.split())
  win_percen = float(pct_tag.get_text(strip=True) or '0')
  # print(win_percen)
  goals_for=t.find('td', class_='gf').get_text(strip=True)
  goals_against=t.find('td', class_='ga').get_text(strip=True)
  
  diff_tag = t.find('td', class_=lambda x: x and 'diff' in str.split(x))
  text_success = float(diff_tag.get_text(strip=True) or '0')
  data.append([name, year, wins, losses, OT_losses, win_percen, goals_for, goals_against, text_success])
  
data
	
```

```python
df_data = pd.DataFrame(data=data, columns=col_name)
df_data
```
